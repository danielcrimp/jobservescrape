```python
import pandas as pd
import re
```

```python
def make_stack_list(descstring):
    #should probably vectorise this(?), yuck
    stacklist = []
    toolslist = ['SQL','Python','Excel','VBA','Power BI','Javascript','Django','React','C#','Scikit-learn','Pandas','Tableau','Azure','AWS','GCP','AI']
    for tool in toolslist:
        if " " + tool.lower()+" " in descstring.lower():
            stacklist.append(tool)
    return stacklist
```

```python
file = pd.read_csv('jobs.csv')
```

```python
df = file.copy(deep=True)
```

```python
df['stacklist'] = df['desc'].apply(make_stack_list)
```

```python
df = df.loc[~df['sal'].isna()]
```

```python
df = df.sort_values(by='stacklist', ascending = False)
```

```python
df['link'] = "https://www.jobserve.com/gb/en/mob/job/" + df['jid']
```

```python
dayfilt1  = df['sal'].str.lower().str.contains('day|daily| pd|p/d')
dayfilt2  = df['sal'].str.lower().str.contains("\d{2,3}pd",regex=True) & ~dayfilt1

dayfilt= dayfilt1 | dayfilt2


hourfilt  = df['sal'].str.lower().str.contains('hour| hr ')

yearfilt1  = df['sal'].str.lower().str.contains('annum|annual|year| yr | pa | p.a | p/a |salary')
yearfilt2  = df['sal'].str.lower().str.contains("\d{2,3}[kK]",regex=True) & ~yearfilt1

yearfilt = yearfilt1 | yearfilt2

df['sal'] = df['sal'].str.replace(',','')
df['sal'] = df['sal'].str.replace("\.00",'',regex=False)

df.loc[yearfilt, 'rate_type'] = "Yearly"
df.loc[dayfilt, 'rate_type'] = "Daily"
df.loc[hourfilt, 'rate_type'] = "Hourly"

```

```python
df['digits'] = df['sal'].apply(lambda x: max(int(i) for i in re.findall('\d+', x)))
df['digits'] = df['digits'].astype(int)
```

```python
df.loc[(yearfilt)&(df['digits']<1000),'digits'] = df['digits']*1000
```

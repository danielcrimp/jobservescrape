```python
import pandas as pd
import re
```

```python
def make_stack_list(descstring):
    #should probably vectorise this(?), yuck
    stacklist = []
    toolslist = ['SQL','Python','Excel','VBA','Power BI','Javascript','Django','React','C#','Scikit-learn','Pandas']
    for tool in toolslist:
        if tool.lower() in descstring.lower():
            stacklist.append(tool)
    return stacklist
```

```python
df = pd.read_csv('jobs.csv')
```

```python
df['stacklist'] = df['desc'].apply(make_stack_list)
```

```python
df = df.loc[~df['sal'].isna()]
```

```python
df = df.sort_values(by='stacklist', ascending = False)
```

```python
df
```

```python
dayfilt  = df['sal'].str.lower().str.contains('day|daily|pd|p/d')
yearfilt  = df['sal'].str.lower().str.contains('annum|annual|year|yr|pa|p.a|salary')

df.loc[yearfilt, 'rate_type'] = "Yearly"
df.loc[dayfilt, 'rate_type'] = "Daily"
```

```python
df['digits'] = df['sal'].apply(lambda x: re.findall('\d+', x))
```
